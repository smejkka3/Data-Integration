```python
#importing modudles
import pandas as pd
import numpy as np
import Levenshtein
import time
import math
```

```python
#loading dataset
df = pd.read_csv('CleanedInputDB.csv')
df = df.apply(lambda x: x.astype(str).str.upper())
```

```python
def string_to_compare(index,what):
    cmp = ''
    cmp = df[what].iloc[index]
    return str(cmp)
```

```python
def cmp_and_append(index0,index1,str0,str1):
    distance = Levenshtein.ratio(str0, str1)
    if(distance > 0.8):
        str0 = string_to_compare(index0,'LastName')
        str1 = string_to_compare(index1,'LastName')
        distance = Levenshtein.ratio(str0, str1)
        if(distance > 0.7):
            str0 = string_to_compare(index0,'FirstName')
            str1 = string_to_compare(index1,'FirstName')
            distance = Levenshtein.ratio(str0, str1)
            if(distance > 0.7):
                return True
    return False
                
```

```python
res=[]
skipped = set()
start_time = time.time()
start_time10 = time.time()
rows = math.ceil(len(df))
for i in range(rows):
    str0 = string_to_compare(i,'SSN')
    if i in skipped:
            continue
    for j in range(i+1,rows):
        str1 = string_to_compare(j,'SSN')
        if(str1 == '0'):
            skipped.add(j)
        if j in skipped:
            continue
        duplicate = cmp_and_append(i,j,str0,str1)
        if(duplicate == True):
            res.append([str(df['RecID'].iloc[i]), str(df['RecID'].iloc[j])])
    if(i%100 == 0 and i != 0):
        print('Passed ', i,' records')
        finish10_time = time.time()
        total10_time = finish10_time - start_time10
        print('avr time for one record so far: ', total10_time/i)
        start_time10 = time.time()
        
finish_time = time.time()
total_time = finish_time - start_time
print('total time', total_time)
```

```python
print('total time: ', total_time/60, ' mins')
```

```python
headers=['tuple_id_1', 'tuple_id_2']

dfout = pd.DataFrame(res, columns=headers)

dfout.to_csv('OutputDB.csv', index=False)
```
